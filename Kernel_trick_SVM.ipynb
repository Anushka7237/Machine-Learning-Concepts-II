```python

from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.datasets._samples_generator import make_circles
import  numpy as np


```

```python
X,y=make_circles(100,noise=.1,factor=.1)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:,0],X[:,1],c=y,cmap='bwr')
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

### when kernel linear is use in SVC

```python
clas=SVC(kernel='linear')
clas.fit(X_train,y_train)
```

```python
y_predict=clas.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_predict)
```

# this is function to plot decision surface 

```python
from matplotlib.colors import ListedColormap
zero_one_colourmap = ListedColormap(('blue', 'red'))
def plot_decision_boundary(X, y, clf):
    X_set, y_set = X, y
    X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, 
                                 stop = X_set[:, 0].max() + 1, 
                                 step = 0.01),
                       np.arange(start = X_set[:, 1].min() - 1, 
                                 stop = X_set[:, 1].max() + 1, 
                                 step = 0.01))
  
    plt.contourf(X1, X2, clf.predict(np.array([X1.ravel(), 
                                             X2.ravel()]).T).reshape(X1.shape),
               alpha = 0.75, 
               cmap = zero_one_colourmap)
    plt.xlim(X1.min(), X1.max())
    plt.ylim(X2.min(), X2.max())
    for i, j in enumerate(np.unique(y_set)):
        plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = (zero_one_colourmap)(i), label = j)
    plt.title('SVM Decision Boundary')
    plt.xlabel('X1')
    plt.ylabel('X2')
    plt.legend()
    return plt.show()
```

```python
import numpy as np
plot_decision_boundary(X, y, clas)
```

### when kernel rbf(Radial Basis Function) is use in SVC

```python
classifier=SVC(kernel='rbf')
classifier.fit(X_train,y_train)
```

```python
y_pred=classifier.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
plot_decision_boundary(X,y,classifier)
```

### when kernel polynomial is use in SVC

```python
clas_new=SVC(kernel='poly',degree=2)
clas_new.fit(X_train,y_train)
```

```python
y_pred_poly=clas_new.predict(X_test)
```

```python
accuracy_score(y_test,y_pred_poly)
```

```python
plot_decision_boundary(X,y,clas_new)
```
