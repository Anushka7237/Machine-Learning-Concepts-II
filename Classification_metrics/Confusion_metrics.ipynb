```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
import pandas as pd
```

```python
df=pd.read_csv('../Datasets/heart_disease_uci.csv')
```

```python
df.shape
```

```python
df['num'].value_counts()
```

```python
df=df.iloc[:,3:]
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df['cp'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df.iloc[:,0:-1],df.iloc[:,-1],test_size=0.2,random_state=42)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe=OneHotEncoder()
```

```python
X_train.head()
```

```python
ohe.fit(X_train[['cp','dataset','restecg']])
```

```python
X_train_new=ohe.transform(X_train[['cp','dataset','restecg']]).toarray()
```

```python
X_test_new=ohe.transform(X_test[['cp','dataset','restecg']]).toarray()
```

```python
clf1=LogisticRegression()
clf2=DecisionTreeClassifier()
```

```python
clf1.fit(X_train_new,y_train)
clf2.fit(X_train_new,y_train)
```

```python
y_predict1=clf1.predict(X_test_new)
y_predict2=clf2.predict(X_test_new)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(f"The Accuracy score after logistic regression is {accuracy_score(y_test,y_predict1)}")
print(f"The Accuracy score after decision tree classfition is {accuracy_score(y_test,y_predict2)}")
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,y_predict1)
```

```python
pd.DataFrame(confusion_matrix(y_test,y_predict2),columns=list(range(0,5)))
```
