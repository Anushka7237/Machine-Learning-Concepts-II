```python
from sklearn.datasets import load_diabetes
import numpy as np
```

```python
X,y=load_diabetes(return_X_y=True)
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from  sklearn.model_selection import cross_val_score
```

```python
lr=LinearRegression()
dt=DecisionTreeRegressor()
svr=SVR()
```

```python
estimators=[('lr',lr),('dt',dt),('svr',svr)]
```

```python
for estimator in estimators:
    score=cross_val_score(estimator[1],X,y,cv=10,scoring='r2')
    print(estimator[0],np.round(np.mean(score),2))
```

```python
from sklearn.ensemble import VotingRegressor
vr=VotingRegressor(estimators=estimators)
score=cross_val_score(vr,X,y,cv=10,scoring='r2')
print(np.round(np.mean(score),2))
```

```python
for i in range(1,4):
    for j in range(1,4):
        for k in range(1,4):
            vr=VotingRegressor(estimators=estimators,weights=[i,j,k])
            x=cross_val_score(vr,X,y,cv=10,scoring='r2')
            print('for i={}, j={} ,k={}'.format(i,j,k),np.round(np.mean(x),2))
```

```python
dt1=DecisionTreeRegressor(max_depth=1)
dt3=DecisionTreeRegressor(max_depth=3)
dt5=DecisionTreeRegressor(max_depth=4)
dt7=DecisionTreeRegressor(max_depth=5)
```

```python
estimators=[('dt1',dt1),('dt3',dt3),('dt5',dt5),('dt7',dt7)]
```

```python
for estimator in estimators:
    score=cross_val_score(estimator[1],X,y,cv=10,scoring='r2')
    print(estimator[0],np.round(np.mean(score),2))
```

```python
vr=VotingRegressor(estimators=estimators)
score=cross_val_score(vr,X,y,cv=10,scoring='r2')
print(np.round(np.mean(score),2))
```
