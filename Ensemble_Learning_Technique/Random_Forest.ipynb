```python
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import seaborn as sns
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv('../Datasets/heart.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
X_train,X_test,y_train,y_test=train_test_split(df.drop(columns='target'),df.iloc[:,-1],test_size=0.2,random_state=42)
```

```python
X_train.head()
```

```python
rf=RandomForestClassifier()
rf.fit(X_train,y_train)
```

```python
y_pred=rf.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
rf=RandomForestClassifier(max_samples=0.75,random_state=42)
rf.fit(X_train,y_train)
```

```python
y_pred=rf.predict(X_test)
accuracy_score(y_test,y_pred)
```

### Tuning Hyperparameters


```python
n_estimators=[20,100,160,200]
max_features=[0.2,0.6,1.0]
max_depth=[2,8,None]
max_samples=[0.5,0.75,1.0]
```

```python
param_grid={
    'n_estimators':n_estimators,
    'max_features':max_features,
    'max_depth':max_depth,
    'max_samples':max_samples
}
print(param_grid)
```

```python
rf_grid=RandomForestClassifier()
```

```python
from sklearn.model_selection import GridSearchCV
search=GridSearchCV(
                estimator=rf_grid,
                param_grid=param_grid,
                cv=5,
                verbose=2,
                n_jobs=-1
)
```

```python
search.fit(X_train,y_train)
```

```python
search.best_params_
```

```python
search.best_score_
```

### RandomSearch cv


```python
n_estimators=[20,100,160,200]
max_features=[0.2,0.6,1.0]
max_depth=[2,8,None]
max_samples=[0.5,0.75,1.0]
bootstrap = [True,False]
min_samples_split = [2, 5]
min_samples_leaf = [1, 2]
```

```python
param_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
              'max_samples':max_samples,
              'bootstrap':bootstrap,
              'min_samples_split':min_samples_split,
              'min_samples_leaf':min_samples_leaf
             }
print(param_grid)
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
search_random=RandomizedSearchCV(
    estimator=rf_grid,
    cv=5,
    param_distributions=param_grid,
    verbose=2,
    n_jobs=-1
)
```

```python
search_random.fit(X_train,y_train)
```

```python
search_random.best_params_
```

```python
search_random.best_score_
```
