```python
from sklearn.datasets import make_regression
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
X,y = make_regression(n_samples=4, n_features=1, n_informative=1, n_targets=1,noise=80,random_state=42)
```

```python
plt.scatter(X,y)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X,y)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X,lr.predict(X),color='red')
```

```python
m=-57.99
b=20
Learning_rate=0.01

epochs=10
for i in range(epochs):
    loss_slope=-2*np.sum(y-m*X.ravel()-b)
    b=b-(Learning_rate*loss_slope)
    
    y_pred=m*X+b
    
    plt.plot(X,y_pred)
plt.scatter(X,y)
```

```python
class GDRegressor:
   
   def __init__(self,learning_rate,epochs):
       self.m=-120
       self.b=-120
       self.learning_rate=learning_rate
       self.epochs=epochs
   def fit(self,X,y):
        for i in range(self.epochs):
            loss_slope_b=-2*(np.sum(y-self.m*X.ravel()-self.b))
            loss_slope_m=-2*(np.sum((y-self.m*X.ravel()-self.b)*X.ravel()))
            self.b=self.b-(self.learning_rate*loss_slope_b)
            self.m=self.m-(self.learning_rate*loss_slope_m)
        print(self.m,self.b)
   def predict(self,X):
        return self.m*X+self.b
```
