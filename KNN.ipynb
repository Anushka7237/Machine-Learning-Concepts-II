```python
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv('Datasets/data.csv')
```

```python
df.head()
```

```python
df.drop(columns=['id','Unnamed: 32'],inplace=True)
```

```python
df.head()
```

```python
X_train,X_test,y_train,y_test=train_test_split(df.drop(columns=['diagnosis']),df['diagnosis'],test_size=0.2,random_state=42)
```

```python
X_train.head()
```

```python
sc=StandardScaler()
sc.fit(X_train)
```

```python
X_train_trans=sc.transform(X_train)
X_test_trans=sc.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model =KNeighborsClassifier(n_neighbors=9)
```

```python
model.fit(X_train_trans,y_train)
```

```python
y_pred=model.predict(X_test_trans)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
for i in range(1,16):
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train_trans,y_train)
    y_predict=knn.predict(X_test_trans)
    
    print(i,accuracy_score(y_test,y_predict))
```
