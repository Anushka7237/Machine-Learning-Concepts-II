```python
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_diabetes
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X,y=load_diabetes(return_X_y=True)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=13)
```

```python
from sklearn.linear_model import SGDRegressor
from sklearn.metrics import r2_score
```

```python
sgd=SGDRegressor(penalty='l2',max_iter=500,eta0=0.1,learning_rate='constant',alpha=0.01)
sgd.fit(X_train,y_train)
y_predict=sgd.predict(X_test)
r2_score(y_test,y_predict)
```

```python
print(sgd.coef_)
```

```python
print(sgd.intercept_)
```

```python
from sklearn.linear_model import Ridge
rig=Ridge(alpha=0.01,max_iter=500,solver='sparse_cg')
```

```python
rig.fit(X_train,y_train)
y_predict1=rig.predict(X_test)
r2_score(y_test,y_predict1)
```

```python
import numpy as np
```

```python
class MyRidge_GD:
    def __init__(self,alpha,learning_rate,epochs):
        self.alpha=alpha
        self.coef_=None
        self.intercept_=None
        self.lr=learning_rate
        self.epochs=epochs
        
    def fit(self,X_train,y_train):
        
        self.coef_=np.ones(X_train.shape[1])
        self.intercept_=0
        theta=np.insert(self.coef_,0,self.intercept_)
        X_train=np.insert(X_train,0,1,axis=1)
        
        for i in range(self.epochs):
            theta_deri=np.dot(X_train.T,X_train).dot(theta)-np.dot(y_train.T,y_train)+self.alpha*theta
            theta=theta-self.lr*theta_deri
            
        self.coef_=theta[1:]
        self.intercept_=theta[0]
        
    def predict(self,X_test):
        return np.dot(X_test,self.coef_)+self.intercept_
        
```

```python
reg = MyRidge_GD(epochs=1000,alpha=0.01,learning_rate=0.005)
```

```python
reg.fit(X_train,y_train)

y_pred = reg.predict(X_test)
print("R2 score",r2_score(y_test,y_pred))
print(reg.coef_)
print(reg.intercept_)
```
