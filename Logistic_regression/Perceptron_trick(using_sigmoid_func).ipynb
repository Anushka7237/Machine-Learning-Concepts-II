```python
from sklearn.datasets import make_classification
import numpy as np
X, y = make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=30)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,4))
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
```

```python
from sklearn.linear_model import LogisticRegression
lor=LogisticRegression()
lor.fit(X,y)
```

```python
m = -(lor.coef_[0][0]/lor.coef_[0][1])
b = -(lor.intercept_/lor.coef_[0][1])
```

```python
X_input=np.linspace(-3,3,100)
Y_input=m*X_input+b
```

```python
plt.figure(figsize=(10,6))
plt.plot(X_input,Y_input,color='black',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```

```python
def check(z):
    return 1 if z>0 else 0
def perceptron(X,y):
    X=np.insert(X,0,1,axis=1)
    weights=np.ones(X.shape[1])
    lr=0.1
    for i in range(1000):
        j=np.random.randint(X.shape[1])
        y_hat=check(np.dot(X[j],weights))
        weights=weights+lr*(y[j]-y_hat)*X[j]
    return weights[0],weights[1:]
```

```python
intercept_,coef_=perceptron(X,y)
```

```python
m=-(coef_[0]/coef_[1])
b=-(intercept_/coef_[1])
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input + b
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))

def perceptron_Sigmoid(X,y):
    X=np.insert(X,0,1,axis=1)
    weights=np.ones(X.shape[1])
    lr=0.1
    
    for i in range(1000):
        j=np.random.randint(X.shape[1])
        y_hat1=sigmoid(np.dot(X[j],weights))
        weights=weights+lr*(y[j]-y_hat1)*X[j]
        
    return weights[0],weights[1:]
```

```python
intercept_sig,coef_sig=perceptron_Sigmoid(X,y)
```

```python
m=-(coef_sig[0]/coef_sig[1])
b=-(intercept_sig/coef_sig[1])
```

```python
X_input_sig=np.linspace(-3,3,100)
Y_input_sig=m*X_input_sig+b
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.plot(X_input,Y_input,color='black',linewidth=3)
plt.plot(X_input_sig,Y_input_sig,color='brown',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```
